```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#  
# image_bgr = cv2.imread('/home/kiwi/github_package/rokeypj_ws/src/turtlebot3_autorace_camera/turtlebot3_autorace_camera/saved_img/saved_image_20250617_200339.jpg')
image_bgr = cv2.imread('/home/kiwi/github_package/rokeypj_ws/src/turtlebot3_autorace_camera/turtlebot3_autorace_camera/saved_img/saved_image_20250618_134300.jpg')
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# BGR  HSV 
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(image_hsv)


#   
print(f'Hue: min={h.min()}, max={h.max()}, mean={h.mean():.2f}')
print(f'Saturation: min={s.min()}, max={s.max()}, mean={s.mean():.2f}')
print(f'Value: min={v.min()}, max={v.max()}, mean={v.mean():.2f}')

#    
plt.figure(figsize=(16, 8))

#  
plt.subplot(2, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Hue 
plt.subplot(2, 3, 2)
plt.hist(h.ravel(), bins=180, range=(0, 180), color='red')
plt.title('Hue Histogram')

# Saturation 
plt.subplot(2, 3, 3)
plt.hist(s.ravel(), bins=256, range=(0, 256), color='green')
plt.title('Saturation Histogram')

# Value 
plt.subplot(2, 3, 4)
plt.hist(v.ravel(), bins=256, range=(0, 256), color='blue')
plt.title('Value Histogram')

# # H  
# plt.subplot(2, 3, 5)
# plt.imshow(h, cmap='hsv')
# plt.title('Hue Channel')
# plt.axis('off')

# # V  
# plt.subplot(2, 3, 6)
# plt.imshow(v, cmap='gray')
# plt.title('Value Channel')
# plt.axis('off')

plt.tight_layout()
plt.show()
```
